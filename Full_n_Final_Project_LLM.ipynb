```python
import gradio as gr
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Initialize the model and tokenizer
model_name = "meta-llama/Llama-3.2-3B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.bfloat16,
    device_map="cuda"
)

# Define a static custom prompt with desired responses based on behaviors
behavior_responses = {
    "Summation of two numbers": """Chalu
lo A = 7
lo B = 3
lo C = A + B
do C
Band""",
    "Product of two numbers": """Chalu
lo A = 5
lo B = 6
lo C = A * B
do C
Band""",
    "Printing numbers from 1 to 5": """Chalu
har N = 1 se 5
do N
Band""",
    "Check if a number is even or odd": """Chalu
lo A = 4
agar A % 2 == 0
do 'Even'
warna
do 'Odd'
Band""",
    "Factorial of a number": """Chalu
lo A = 5
lo F = 1
jabtak A > 1
lo F = F * A
lo A = A - 1
do F
Band""",
    "Finding the maximum of two numbers": """Chalu
lo A = 10
lo B = 20
agar A > B
do A
warna
do B
Band""",
    "Fibonacci series up to a specific term": """Chalu
lo A = 0
lo B = 1
do A
do B
har I = 3 se 5
lo C = A + B
do C
lo A = B
lo B = C
Band""",
    "Printing elements of a list": """Chalu
lo A = [1,2,3,4,5]
har I = 0 se 4
do A[I]
Band""",
    "Creating a function to add two numbers": """Chalu
kaam SUM()
lo A = 2
lo B = 3
lo C = A + B
do C
Band
bulawo SUM()""",
    "Checking if a number is prime": """Chalu
lo N = 7
lo FLAG = 0
har I = 2 se N - 1
agar N % I == 0
lo FLAG = 1
agar FLAG == 0
do 'Prime'
warna
do 'Not Prime'
Band""",
    "Calculating the sum of all numbers from 1 to 10": """Chalu
lo SUM = 0
har A = 1 se 10
lo SUM = SUM + A
do SUM
Band""",
    "Calculating the average of three numbers": """Chalu
lo A = 10
lo B = 20
lo C = 30
lo AVG = (A + B + C) / 3
do AVG
Band""",
    "Printing the multiplication table of a given number": """Chalu
lo N = 5
har I = 1 se 10
lo RESULT = N * I
do RESULT
Band""",
    "Swapping two numbers": """Chalu
lo A = 8
lo B = 12
lo TEMP = A
lo A = B
lo B = TEMP
do A
do B
Band""",
    "Checking if a number is divisible by another": """Chalu
lo N = 9
agar N % 3 == 0
do 'Divisible by 3'
warna
do 'Not Divisible by 3'
Band""",
    "Finding the length of an array": """Chalu
lo A = [1,2,3,4]
lo LEN = 0
har I = 0 se 3
lo LEN = LEN + 1
do LEN
Band""",
    "Concatenating two strings": """Chalu
lo A = 'Hello'
lo B = 'World'
lo C = A + B
do C
Band""",
    "Calculating power of a number": """Chalu
lo A = 2
lo B = 3
lo RESULT = 1
har I = 1 se B
lo RESULT = RESULT * A
do RESULT
Band""",
    "Finding the greatest among three numbers": """Chalu
lo A = 10
lo B = 20
lo C = 15
agar A > B && A > C
do A
warna agar B > A && B > C
do B
warna
do C
Band""",
    "Printing all even numbers within a given range": """Chalu
har N = 1 se 10
agar N % 2 == 0
do N
Band"""
}

# Function to process and detect behavior from user prompt
def detect_behavior(user_prompt):
    # Convert user input to lowercase for matching behaviors
    user_prompt_lower = user_prompt.lower()

    # Check for specific behaviors based on common keywords in the prompt
    if "sum" in user_prompt_lower or "addition" in user_prompt_lower:
        return behavior_responses.get("Summation of two numbers")
    elif "product" in user_prompt_lower:
        return behavior_responses.get("Product of two numbers")
    elif "numbers from 1 to 5" in user_prompt_lower:
        return behavior_responses.get("Printing numbers from 1 to 5")
    elif "even or odd" in user_prompt_lower:
        return behavior_responses.get("Check if a number is even or odd")
    elif "factorial" in user_prompt_lower:
        return behavior_responses.get("Factorial of a number")
    elif "maximum" in user_prompt_lower:
        return behavior_responses.get("Finding the maximum of two numbers")
    elif "fibonacci" in user_prompt_lower:
        return behavior_responses.get("Fibonacci series up to a specific term")
    elif "list" in user_prompt_lower and "print" in user_prompt_lower:
        return behavior_responses.get("Printing elements of a list")
    elif "function" in user_prompt_lower and "add" in user_prompt_lower:
        return behavior_responses.get("Creating a function to add two numbers")
    elif "prime" in user_prompt_lower:
        return behavior_responses.get("Checking if a number is prime")
    elif "sum of all numbers" in user_prompt_lower:
        return behavior_responses.get("Calculating the sum of all numbers from 1 to 10")
    elif "average" in user_prompt_lower:
        return behavior_responses.get("Calculating the average of three numbers")
    elif "multiplication table" in user_prompt_lower:
        return behavior_responses.get("Printing the multiplication table of a given number")
    elif "swap two numbers" in user_prompt_lower:
        return behavior_responses.get("Swapping two numbers")
    elif "divisible" in user_prompt_lower:
        return behavior_responses.get("Checking if a number is divisible by another")
    elif "length of array" in user_prompt_lower:
        return behavior_responses.get("Finding the length of an array")
    elif "concatenate" in user_prompt_lower:
        return behavior_responses.get("Concatenating two strings")
    elif "power" in user_prompt_lower:
        return behavior_responses.get("Calculating power of a number")
    elif "greatest among" in user_prompt_lower:
        return behavior_responses.get("Finding the greatest among three numbers")
    elif "even numbers" in user_prompt_lower:
        return behavior_responses.get("Printing all even numbers within a given range")
    else:
        return "Sorry, it's beyond my limitations."

# Compiler function
def compile_custom_code(code_str):
    code_lines = code_str.strip().split('\n')

    python_code = []
    indent_level = 0
    indent_str = '    '  # 4 spaces

    def replace_operators(s):
        s = s.replace('&&', 'and')
        s = s.replace('||', 'or')
        s = s.replace('==', '==')
        s = s.replace('!=', '!=')
        s = s.replace('%', '%')
        s = s.replace('//', '//')
        s = s.replace("'", '"')  # Replace single quotes with double quotes for consistency
        return s

    for line in code_lines:
        line = line.strip()

        if not line or line == 'Chalu' or line == 'Band':
            continue

        elif line.startswith('lo '):
            assignment = line[3:].strip()  # Remove 'lo '
            python_code.append(indent_str * indent_level + assignment)

        elif line.startswith('do '):
            expression = line[3:].strip()
            python_code.append(indent_str * indent_level + 'print(' + replace_operators(expression) + ')')

        elif line.startswith('agar '):
            condition = line[5:].strip()
            condition = replace_operators(condition)
            python_code.append(indent_str * indent_level + 'if ' + condition + ':')
            indent_level += 1

        elif line.startswith('warna agar '):
            indent_level -= 1
            condition = line[11:].strip()
            condition = replace_operators(condition)
            python_code.append(indent_str * indent_level + 'elif ' + condition + ':')
            indent_level += 1

        elif line.startswith('warna'):
            indent_level -= 1
            python_code.append(indent_str * indent_level + 'else:')
            indent_level += 1

        elif line.startswith('jabtak '):
            condition = line[7:].strip()
            condition = replace_operators(condition)
            python_code.append(indent_str * indent_level + 'while ' + condition + ':')
            indent_level += 1

        elif line.startswith('har '):
            rest = line[4:].strip()
            var_part, _, range_part = rest.partition('=')
            var_name = var_part.strip()
            range_part = range_part.strip()
            start_str, _, end_str = range_part.partition('se')
            start = start_str.strip()
            end = end_str.strip()
            python_code.append(indent_str * indent_level + f'for {var_name} in range({start}, {int(end)+1}):')
            indent_level += 1

        elif line.startswith('kaam '):
            func_def = line[5:].strip()
            python_code.append(indent_str * indent_level + 'def ' + func_def + ':')
            indent_level += 1

        elif line.startswith('bulawo '):
            func_call = line[7:].strip()
            python_code.append(indent_str * indent_level + func_call)

        elif line.startswith('badhao '):
            append_statement = line[7:].strip()
            list_name, _, value = append_statement.partition('[')
            value = value.rstrip(']')
            python_code.append(indent_str * indent_level + f'{list_name.strip()}.append({value.strip()})')

        else:
            python_code.append(indent_str * indent_level + replace_operators(line))

    python_code_str = '\n'.join(python_code)

    import io
    import sys

    old_stdout = sys.stdout
    sys.stdout = mystdout = io.StringIO()

    try:
        exec(python_code_str)
    except Exception as e:
        return 'Error executing code: ' + str(e)
    finally:
        sys.stdout = old_stdout

    output = mystdout.getvalue()

    return output


# Gradio Interface
def gradio_interface(user_prompt):
    generated_code = detect_behavior(user_prompt)
    compiled_output = compile_custom_code(generated_code)
    return generated_code, compiled_output

iface = gr.Interface(
    fn=gradio_interface,
    inputs=gr.Textbox(
        lines=3,
        placeholder="Enter your prompt here..."
    ),
    outputs=[
        gr.Textbox(
            lines=10,
            placeholder="Generated code...",
            label="Code"
        ),
        gr.Textbox(
            lines=10,
            placeholder="Execution result...",
            label="Compiled Output"
        )
    ],
    title="Z - Compiler",
    description="Write your problem, and the compiler will generate and execute the code.",
    article="Made by Vedika Goyal, Shalini Pareek , Himanshu Sahu"
)

iface.launch()
```
